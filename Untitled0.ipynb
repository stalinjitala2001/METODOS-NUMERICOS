b. Use el método de bisección para encontrar una aproximación dentro de 10ିହ para el primer valor positivo
de 𝑥 con 𝑥 = tan 𝑥.


````python
import numpy as np

def f(x):
    return np.tan(x) - x

def bisection_method(a, b, tol):
    while (b - a) / 2.0 > tol:
        midpoint = (a + b) / 2.0
        if f(midpoint) == 0:
            return midpoint
        elif f(a) * f(midpoint) < 0:
            b = midpoint
        else:
            a = midpoint
    return (a + b) / 2.0

# Definir los límites del intervalo y la tolerancia
a = 0
b = np.pi / 2
tol = 1e-3

# Calcular la aproximación de la raíz usando el método de bisección
approx_root = bisection_method(a, b, tol)

print("Aproximación de la raíz:", approx_root)
````

Un abrevadero de longitud L tiene una sección transversal en forma de semicírculo con radio r. Cuando se llena con agua hasta una distancia h a partir de la parte superior, el volumen V de agua es
V=L[0.5\pir^2-r^2arcsen(\frac{h}{r})-h\left(r^2-h^2\right)^\frac{1}{2}
Suponga que 𝐿 = 10 𝑐𝑚, 𝑟 = 1 𝑐𝑚 y 𝑉 = 12.4 𝑐𝑚ଷ . Encuentre la profundidad del agua en el abrevadero dentro de 0.01 𝑐𝑚.


````python
import numpy as np

# Definimos los parámetros
L = 10  # longitud en cm
r = 1  # radio en cm
V = 12.4  # volumen en cm³

# Definimos la función f(h)
def f(h):
    term1 = 0.5 * np.pi * r**2
    term2 = r**2 * np.arcsin(h / r)
    term3 = h * np.sqrt(r**2 - h**2)
    return L * (term1 - term2 - term3) - V

# Método de bisección
def bisection_method(func, a, b, tol):
    if func(a) * func(b) > 0:
        print("El método de bisección no puede garantizar la convergencia en este intervalo.")
        return None
    else:
        while (b - a) / 2.0 > tol:
            midpoint = (a + b) / 2.0
            if func(midpoint) == 0:
                return midpoint
            elif func(a) * func(midpoint) < 0:
                b = midpoint
            else:
                a = midpoint
        return (a + b) / 2.0

# Intervalo inicial para h (de 0 a r)
a = 0
b = r

# Tolerancia
tol = 0.01

# Ejecutar el método de bisección
h = bisection_method(f, a, b, tol)

if h is not None:
    print("La profundidad del agua en el abrevadero es aproximadamente:", h, "cm")
else:
    print("No se pudo encontrar una profundidad en el intervalo dado con el método de bisección.")
````

Un objeto que cae verticalmente a través del aire está sujeto a una resistencia viscosa, así como a la fuerza de gravedad. Suponga que un objeto con masa 𝑚 cae desde una altura 𝑠଴ y que la altura del objeto después de 𝑡 segundos es
s\left(t\right)=s_0-\frac{mg}{k}t+\frac{m^2g}{k^2}\left(1-e^\frac{-kt}{m}\right)
donde 𝑔 = 9.81 𝑚 𝑠ൗ ଶ y 𝑘 representa el coeficiente de la resistencia del aire en 𝑁𝑠ൗ𝑚. Suponga 𝑠଴ = 300 𝑚, 𝑚 = 0.25 𝑘𝑔 y 𝑘 = 0.1 𝑁𝑠ൗ𝑚. Encuentre, dentro de 0.01 𝑠𝑒𝑔𝑢𝑛𝑑𝑜𝑠, el tiempo que tarda un cuarto de kg en golpear el piso.


````python
import numpy as np

# Definir los parámetros
s0 = 300  # altura inicial en metros
m = 0.25  # masa en kg
g = 9.81  # aceleración debido a la gravedad en m/s^2
k = 0.1   # coeficiente de resistencia del aire en Ns/m

# Definir la función f(t)
def f(t):
    term1 = s0
    term2 = (m * g / k) * t
    term3 = (m**2 * g / k**2) * (1 - np.exp(-k * t / m))
    return term1 - term2 + term3

# Método de bisección
def bisection_method(func, a, b, tol):
    if func(a) * func(b) > 0:
        print("El método de bisección no puede garantizar la convergencia en este intervalo.")
        return None
    else:
        while (b - a) / 2.0 > tol:
            midpoint = (a + b) / 2.0
            if func(midpoint) == 0:
                return midpoint
            elif func(a) * func(midpoint) < 0:
                b = midpoint
            else:
                a = midpoint
        return (a + b) / 2.0

# Intervalo inicial para t (de 0 a un valor grande)
a = 0
b = 100  # suposición inicial grande para el intervalo

# Tolerancia
tol = 0.01

# Ejecutar el método de bisección
t = bisection_method(f, a, b, tol)

if t is not None:
    print("El tiempo que tarda en golpear el piso es aproximadamente:", t, "segundos")
else:
    print("No se pudo encontrar el tiempo en el intervalo dado con el método de bisección.")
````

Use el teorema 2.1 para encontrar una cota para el número de iteraciones necesarias para lograr una
aproximación con precisión de 10ିସ para la solución de 𝑥
ଷ − 𝑥 − 1 = 0 que se encuentra dentro del intervalo
[1, 2]. Encuentre una aproximación para la raíz con este grado de precisión.


````python
import math

# Definir la función
def f(x):
    return x**3 - x - 1

# Método de bisección
def bisection_method(f, a, b, tol):
    if f(a) * f(b) > 0:
        raise ValueError("El método de bisección no garantiza una raíz en este intervalo.")

    iterations = 0
    while (b - a) / 2.0 > tol:
        midpoint = (a + b) / 2.0
        if f(midpoint) == 0:
            return midpoint, iterations
        elif f(a) * f(midpoint) < 0:
            b = midpoint
        else:
            a = midpoint
        iterations += 1

    return (a + b) / 2.0, iterations

# Intervalo inicial
a = 1
b = 2
# Tolerancia
tol = 10**-4

# Ejecutar el método de bisección
root, num_iterations = bisection_method(f, a, b, tol)

print(f"La raíz aproximada es: {root}")
print(f"Número de iteraciones: {num_iterations}")
````

````python
import math

def f(x):
    return math.sin(math.pi * x)

def bisection_method(f, a, b, tol):
    if f(a) * f(b) > 0:
        raise ValueError("El método de bisección no garantiza una raíz en este intervalo.")

    iterations = 0
    while (b - a) / 2.0 > tol:
        midpoint = (a + b) / 2.0
        if f(midpoint) == 0:
            return midpoint, iterations
        elif f(a) * f(midpoint) < 0:
            b = midpoint
        else:
            a = midpoint
        iterations += 1

    return (a + b) / 2.0, iterations

# Definir los intervalos y la tolerancia
intervals = [(-1, 0), (2, 3)]
tolerance = 10**-5

# Ejecutar el método de bisección para ambos intervalos
results = {}
for interval in intervals:
    a, b = interval
    root, iterations = bisection_method(f, a, b, tolerance)
    results[interval] = (root, iterations)

# Imprimir resultados
for interval, result in results.items():
    root, iterations = result
    print(f"Intervalo: {interval}")
    print(f"Raíz aproximada: {root}")
    print(f"Número de iteraciones: {iterations}")
    print()
````
