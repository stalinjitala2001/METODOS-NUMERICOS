b. Use el mÃ©todo de bisecciÃ³n para encontrar una aproximaciÃ³n dentro de 10à¬¿à¬¹ para el primer valor positivo
de ð‘¥ con ð‘¥ = tan ð‘¥.


````python
import numpy as np

def f(x):
    return np.tan(x) - x

def bisection_method(a, b, tol):
    while (b - a) / 2.0 > tol:
        midpoint = (a + b) / 2.0
        if f(midpoint) == 0:
            return midpoint
        elif f(a) * f(midpoint) < 0:
            b = midpoint
        else:
            a = midpoint
    return (a + b) / 2.0

# Definir los lÃ­mites del intervalo y la tolerancia
a = 0
b = np.pi / 2
tol = 1e-3

# Calcular la aproximaciÃ³n de la raÃ­z usando el mÃ©todo de bisecciÃ³n
approx_root = bisection_method(a, b, tol)

print("AproximaciÃ³n de la raÃ­z:", approx_root)
````

Un abrevadero de longitud L tiene una secciÃ³n transversal en forma de semicÃ­rculo con radio r. Cuando se llena con agua hasta una distancia h a partir de la parte superior, el volumen V de agua es
V=L[0.5\pir^2-r^2arcsen(\frac{h}{r})-h\left(r^2-h^2\right)^\frac{1}{2}
Suponga que ð¿ = 10 ð‘ð‘š, ð‘Ÿ = 1 ð‘ð‘š y ð‘‰ = 12.4 ð‘ð‘šà¬· . Encuentre la profundidad del agua en el abrevadero dentro de 0.01 ð‘ð‘š.


````python
import numpy as np

# Definimos los parÃ¡metros
L = 10  # longitud en cm
r = 1  # radio en cm
V = 12.4  # volumen en cmÂ³

# Definimos la funciÃ³n f(h)
def f(h):
    term1 = 0.5 * np.pi * r**2
    term2 = r**2 * np.arcsin(h / r)
    term3 = h * np.sqrt(r**2 - h**2)
    return L * (term1 - term2 - term3) - V

# MÃ©todo de bisecciÃ³n
def bisection_method(func, a, b, tol):
    if func(a) * func(b) > 0:
        print("El mÃ©todo de bisecciÃ³n no puede garantizar la convergencia en este intervalo.")
        return None
    else:
        while (b - a) / 2.0 > tol:
            midpoint = (a + b) / 2.0
            if func(midpoint) == 0:
                return midpoint
            elif func(a) * func(midpoint) < 0:
                b = midpoint
            else:
                a = midpoint
        return (a + b) / 2.0

# Intervalo inicial para h (de 0 a r)
a = 0
b = r

# Tolerancia
tol = 0.01

# Ejecutar el mÃ©todo de bisecciÃ³n
h = bisection_method(f, a, b, tol)

if h is not None:
    print("La profundidad del agua en el abrevadero es aproximadamente:", h, "cm")
else:
    print("No se pudo encontrar una profundidad en el intervalo dado con el mÃ©todo de bisecciÃ³n.")
````

Un objeto que cae verticalmente a travÃ©s del aire estÃ¡ sujeto a una resistencia viscosa, asÃ­ como a la fuerza de gravedad. Suponga que un objeto con masa ð‘š cae desde una altura ð‘ à¬´ y que la altura del objeto despuÃ©s de ð‘¡ segundos es
s\left(t\right)=s_0-\frac{mg}{k}t+\frac{m^2g}{k^2}\left(1-e^\frac{-kt}{m}\right)
donde ð‘” = 9.81 ð‘š ð‘ àµ— à¬¶ y ð‘˜ representa el coeficiente de la resistencia del aire en ð‘ð‘ àµ—ð‘š. Suponga ð‘ à¬´ = 300 ð‘š, ð‘š = 0.25 ð‘˜ð‘” y ð‘˜ = 0.1 ð‘ð‘ àµ—ð‘š. Encuentre, dentro de 0.01 ð‘ ð‘’ð‘”ð‘¢ð‘›ð‘‘ð‘œð‘ , el tiempo que tarda un cuarto de kg en golpear el piso.


````python
import numpy as np

# Definir los parÃ¡metros
s0 = 300  # altura inicial en metros
m = 0.25  # masa en kg
g = 9.81  # aceleraciÃ³n debido a la gravedad en m/s^2
k = 0.1   # coeficiente de resistencia del aire en Ns/m

# Definir la funciÃ³n f(t)
def f(t):
    term1 = s0
    term2 = (m * g / k) * t
    term3 = (m**2 * g / k**2) * (1 - np.exp(-k * t / m))
    return term1 - term2 + term3

# MÃ©todo de bisecciÃ³n
def bisection_method(func, a, b, tol):
    if func(a) * func(b) > 0:
        print("El mÃ©todo de bisecciÃ³n no puede garantizar la convergencia en este intervalo.")
        return None
    else:
        while (b - a) / 2.0 > tol:
            midpoint = (a + b) / 2.0
            if func(midpoint) == 0:
                return midpoint
            elif func(a) * func(midpoint) < 0:
                b = midpoint
            else:
                a = midpoint
        return (a + b) / 2.0

# Intervalo inicial para t (de 0 a un valor grande)
a = 0
b = 100  # suposiciÃ³n inicial grande para el intervalo

# Tolerancia
tol = 0.01

# Ejecutar el mÃ©todo de bisecciÃ³n
t = bisection_method(f, a, b, tol)

if t is not None:
    print("El tiempo que tarda en golpear el piso es aproximadamente:", t, "segundos")
else:
    print("No se pudo encontrar el tiempo en el intervalo dado con el mÃ©todo de bisecciÃ³n.")
````

Use el teorema 2.1 para encontrar una cota para el nÃºmero de iteraciones necesarias para lograr una
aproximaciÃ³n con precisiÃ³n de 10à¬¿à¬¸ para la soluciÃ³n de ð‘¥
à¬· âˆ’ ð‘¥ âˆ’ 1 = 0 que se encuentra dentro del intervalo
[1, 2]. Encuentre una aproximaciÃ³n para la raÃ­z con este grado de precisiÃ³n.


````python
import math

# Definir la funciÃ³n
def f(x):
    return x**3 - x - 1

# MÃ©todo de bisecciÃ³n
def bisection_method(f, a, b, tol):
    if f(a) * f(b) > 0:
        raise ValueError("El mÃ©todo de bisecciÃ³n no garantiza una raÃ­z en este intervalo.")

    iterations = 0
    while (b - a) / 2.0 > tol:
        midpoint = (a + b) / 2.0
        if f(midpoint) == 0:
            return midpoint, iterations
        elif f(a) * f(midpoint) < 0:
            b = midpoint
        else:
            a = midpoint
        iterations += 1

    return (a + b) / 2.0, iterations

# Intervalo inicial
a = 1
b = 2
# Tolerancia
tol = 10**-4

# Ejecutar el mÃ©todo de bisecciÃ³n
root, num_iterations = bisection_method(f, a, b, tol)

print(f"La raÃ­z aproximada es: {root}")
print(f"NÃºmero de iteraciones: {num_iterations}")
````

````python
import math

def f(x):
    return math.sin(math.pi * x)

def bisection_method(f, a, b, tol):
    if f(a) * f(b) > 0:
        raise ValueError("El mÃ©todo de bisecciÃ³n no garantiza una raÃ­z en este intervalo.")

    iterations = 0
    while (b - a) / 2.0 > tol:
        midpoint = (a + b) / 2.0
        if f(midpoint) == 0:
            return midpoint, iterations
        elif f(a) * f(midpoint) < 0:
            b = midpoint
        else:
            a = midpoint
        iterations += 1

    return (a + b) / 2.0, iterations

# Definir los intervalos y la tolerancia
intervals = [(-1, 0), (2, 3)]
tolerance = 10**-5

# Ejecutar el mÃ©todo de bisecciÃ³n para ambos intervalos
results = {}
for interval in intervals:
    a, b = interval
    root, iterations = bisection_method(f, a, b, tolerance)
    results[interval] = (root, iterations)

# Imprimir resultados
for interval, result in results.items():
    root, iterations = result
    print(f"Intervalo: {interval}")
    print(f"RaÃ­z aproximada: {root}")
    print(f"NÃºmero de iteraciones: {iterations}")
    print()
````
